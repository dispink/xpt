```python
import folium

# Create a map centered at a specific location
m = folium.Map(location=[2.71666,	34.06666], zoom_start=5)

# List of coordinates to plot
coordinates = [
    [33.395,	33.30166],
    [31.79833,	33.03166],
    [31.65,	32.345],
    [30.03333,	42.93333],
    [32.71666,	34.06666],
    [4.96666,	52.38333],
    [4.98333,	52.33333],
    [3.71666,	51.7],
    [4.3,	51.76666]
]

# Add points to the map
for coord in coordinates:
    folium.Marker(location=coord).add_to(m)

# Display the map
m
```
